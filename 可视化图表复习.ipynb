```python
# matplotlib 的图像都位于 Figure 对象中。可以用 plt.figure 创建一个新的Figure
fig = plt.figure()
```

```python
#不能通过空 Figure 绘图。必须用 add_subplot 创建一个或多个 subplot 才行
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
#如果没有指定在哪个图画，则默认在最后一个图，比如plt.plot(np.random.randn(50).cumsum(),'k--')
#randn(50)在标准正态分布中返回50个值
ax3.plot(np.random.randn(50).cumsum(),'k--')
ax1.hist(np.random.randn(100),bins=20,color='k',alpha=0.3)
ax2.scatter(np.arange(30),np.arange(30)+3*np.random.randn(30))
```

```python
#sharex=True,sharey=True。所有subplot使用相同的X轴Y轴刻度，rand生成的数字在0-1之间，bins=50生成50条柱
fig,axes = plt.subplots(2,2,sharex=True,sharey=True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.rand(500),bins=50,color='k',alpha=0.5)
#每个图之间的距离为0
plt.subplots_adjust(wspace=0,hspace=0)
```

```python
fig,axes = plt.subplots(2,2)
x = np.arange(10)
y = np.random.randn(10)
#'g--'相当于linestyle='--',color='g'
axes[0,1].plot(x,y,'g--')
```

```python
#重新创建一个视图
fig = plt.figure()
x = np.arange(10)
y = np.random.randn(10)
plt.plot(x,y,'g--')
```

```python
fig = plt.figure()
#'ko--'相当于color='k',linestyle='--',marker='o'
plt.plot(np.random.randn(30).cumsum(),'ko--',label='Default')
plt.plot(np.random.randn(30).cumsum(),'ko-',label='steps-post',drawstyle='steps-post')
plt.legend(loc='best')
```

```python
#下两行即等价于fig,ax = plt.subplots()
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.plot(np.random.randn(1000).cumsum())
ticks = ax.set_xticks([0,250,500,750,1000])
labels = ax.set_xticklabels(['one','two','three','four','five'],rotation=30,fontsize='small')
ax.set_title('My fist matplotlib plot')
ax.set_xlabel('Stages')
# plt.savefig('My fist matplotlib plot.png')

#批量设定绘图选项
# props = {
#     'title': 'My first matplotlib plot',
#     'xlabel': 'Stages' 
# } 
# ax.set(**props)

#设定xy轴范围
# ax.set_xlim([1, 10])
# ax.set_ylim([1, 10])
```

```python
#添加图例
fig, axes = plt.subplots()
axes.plot(np.random.randn(100).cumsum(), 'k', label='one')
axes.plot(np.random.randn(100).cumsum(), 'k--', label='two')
axes.plot(np.random.randn(100).cumsum(), 'k.', label='three')
axes.legend(loc='best')
```
