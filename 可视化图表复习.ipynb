```python
#ä¸èƒ½é€šè¿‡ç©º Figure ç»˜å›¾ã€‚å¿…é¡»ç”¨ add_subplot åˆ›å»ºä¸€ä¸ªæˆ–å¤šä¸ª subplot æ‰è¡Œ
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
#å¦‚æžœæ²¡æœ‰æŒ‡å®šåœ¨å“ªä¸ªå›¾ç”»ï¼Œåˆ™é»˜è®¤åœ¨æœ€åŽä¸€ä¸ªå›¾ï¼Œæ¯”å¦‚plt.plot(np.random.randn(50).cumsum(),'k--')
#randn(50)åœ¨æ ‡å‡†æ­£æ€åˆ†å¸ƒä¸­è¿”å›ž50ä¸ªå€¼
ax3.plot(np.random.randn(50).cumsum(),'k--')
ax1.hist(np.random.randn(100),bins=20,color='k',alpha=0.3)
ax2.scatter(np.arange(30),np.arange(30)+3*np.random.randn(30))
```

```python
#sharex=True,sharey=Trueã€‚æ‰€æœ‰subplotä½¿ç”¨ç›¸åŒçš„Xè½´Yè½´åˆ»åº¦ï¼Œrandç”Ÿæˆçš„æ•°å­—åœ¨0-1ä¹‹é—´ï¼Œbins=50ç”Ÿæˆ50æ¡æŸ±
fig,axes = plt.subplots(2,2,sharex=True,sharey=True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.rand(500),bins=50,color='k',alpha=0.5)
#æ¯ä¸ªå›¾ä¹‹é—´çš„è·ç¦»ä¸º0
plt.subplots_adjust(wspace=0,hspace=0)
```
